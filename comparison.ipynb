```python
def mean_rouge(di_sum, oth):
  sred = []
  sr1 = 0
  sr2 = 0
  sl = 0
  s1 = 0
  s2 = 0
  s3 = 0
  s4 = 0
  for i in range(len(di_sum)):
    sr1 += di_sum[i]['rouge1'].fmeasure
    sr2 += di_sum[i]['rouge2'].fmeasure
    sl += di_sum[i]['rougeL'].fmeasure
    s1 += oth[i]['rouge1']
    s2 += oth[i]['rouge2']
    s3 += oth[i]['rougeL']
    s4 += oth[i]['rougeLsum']
  sr1 = sr1 / len(di_sum)
  sr2 = sr2 / len(di_sum)
  sl = sl / len(di_sum)
  s1 = s1 / len(oth)
  s2 = s2 / len(oth)
  s3 = s3 / len(oth)
  s4 = s4 / len(oth)
  return sr1, sr2, sl, s1, s2, s3, s4
```
